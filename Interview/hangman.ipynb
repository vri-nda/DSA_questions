```python
class HangmanAPI(object):
    def __init__(self, access_token=None, session=None, timeout=None):
        self.hangman_url = self.determine_hangman_url()
        self.access_token = access_token
        self.session = session or requests.Session()
        self.timeout = timeout
        self.guessed_letters = []
        
        full_dictionary_location = "words_250000_train.txt"
        self.full_dictionary = self.build_dictionary(full_dictionary_location)        
        self.full_dictionary_common_letter_sorted = collections.Counter("".join(self.full_dictionary)).most_common()
        self.current_dictionary = []
        
        
    @staticmethod
    def determine_hangman_url():
        links = ['https://trexsim.com', 'https://sg.trexsim.com']

        data = {link: 0 for link in links}

        for link in links:

            requests.get(link)

            for i in range(10):
                s = time.time()
                requests.get(link)
                data[link] = time.time() - s

        link = sorted(data.items(), key=lambda x: x[1])[0][0]
        link += '/trexsim/hangman'
        return link

    def guess(self, word): # word input example: "_ p p _ e "
        ###############################################
        # Replace with your own "guess" function here #
        ###############################################
        
        ############################################### guess function 1  efficiency 0.23-0.26 tak ki aa rahi
        # clean_word = word[::2].replace("_", ".")
        # len_word = len(clean_word)
        # current_dictionary = self.current_dictionary
        # new_dictionary = []

        # for dict_word in current_dictionary:
        #     if len(dict_word) != len_word:
        #         continue
        #     if re.match(clean_word, dict_word):
        #         new_dictionary.append(dict_word)

        # self.current_dictionary = new_dictionary
        # full_dict_string = "".join(new_dictionary)
        # c = collections.Counter(full_dict_string)
        # sorted_letter_count = c.most_common()
        
        # guess_letter = '!'
        # for letter, instance_count in sorted_letter_count:
        #     if letter not in self.guessed_letters:
        #         guess_letter = letter
        #         break

        # if guess_letter == '!':
        #     sorted_letter_count = self.full_dictionary_common_letter_sorted
        #     for letter, instance_count in sorted_letter_count:
        #         if letter not in self.guessed_letters:
        #             guess_letter = letter
        #             break

        # return guess_letter
        
        ############################################################################################################################################# 
      
        #############################################################################################################################################
        
        
        ############################################### guess function 3  efficiency 12.1%
        # clean_word = word[::2].replace("_", ".")
        # len_word = len(clean_word)
        # current_dictionary = self.current_dictionary
        # new_dictionary = []

        # for dict_word in current_dictionary:
        #     if len(dict_word) != len_word:
        #         continue
        #     if re.match(clean_word, dict_word):
        #         new_dictionary.append(dict_word)

        # self.current_dictionary = new_dictionary
        # full_dict_string = "".join(new_dictionary)
        # c = collections.Counter(full_dict_string)
        # sorted_letter_count = c.most_common()
        
        # # Calculate letter frequencies in the English language
        # english_letter_frequencies = {
        #     'e': 12.702, 't': 9.056, 'a': 8.167, 'o': 7.507, 'i': 6.966,
        #     'n': 6.749, 's': 6.327, 'h': 6.094, 'r': 5.987, 'd': 4.253,
        #     'l': 4.025, 'c': 2.782, 'u': 2.758, 'm': 2.406, 'w': 2.361,
        #     'f': 2.228, 'g': 2.015, 'y': 1.974, 'p': 1.929, 'b': 1.492,
        #     'v': 0.978, 'k': 0.772, 'j': 0.153, 'x': 0.150, 'q': 0.095,
        #     'z': 0.074
        # }
        
        # # Find the most common letter in the possible words that hasn't been guessed yet
        # guess_letter = '!'
        # for letter, instance_count in sorted_letter_count:
        #     if letter not in self.guessed_letters:
        #         guess_letter = letter
        #         break
        
        # if guess_letter == '!':
        #     # If no word matches in the training dictionary, default back to ordering of full dictionary
        #     sorted_letter_count = self.full_dictionary_common_letter_sorted
        #     for letter, instance_count in sorted_letter_count:
        #         if letter not in self.guessed_letters:
        #             guess_letter = letter
        #             break
        #     # Check if we can improve the guess based on English letter frequencies
        #     if len_word > 1:
        #         english_freq_guesses = []
        #         for letter, instance_count in sorted_letter_count:
        #             if letter not in self.guessed_letters:
        #                 english_freq_guesses.append((letter, english_letter_frequencies[letter]))
        #         if english_freq_guesses:
        #             guess_letter = max(english_freq_guesses, key=lambda x: x[1])[0]
            
        # # Improvements
        # if len_word > 1 and '_' in word:
        #     # Prioritize guessing vowels if word pattern has initial blanks
        #     initial_blanks = word.startswith("_")
        #     vowels = set('aeiou')
        #     if initial_blanks:
        #         available_vowels = [letter for letter in english_letter_frequencies if letter in vowels and letter not in self.guessed_letters]
        #         if available_vowels:
        #             guess_letter = max(available_vowels, key=lambda x: english_letter_frequencies[x])

        # # Consider word frequency if no strong clues are available
        # if guess_letter == '!' and len_word > 1:
        #     word_freq_guesses = []
        #     for letter, instance_count in sorted_letter_count:
        #         if letter not in self.guessed_letters:
        #             word_freq_guesses.append((letter, self.get_letter_word_frequency(letter)))
        #     if word_freq_guesses:
        #         guess_letter = max(word_freq_guesses, key=lambda x: x[1])[0]

        # return guess_letter
        #############################################################################################################################################
        
        


        
        
        
        
        
        ############################################### function 4 based on ML algorith
    #     clean_word = word[::2].replace("_", ".")
    #     len_word = len(clean_word)
    #     current_dictionary = self.current_dictionary
    #     new_dictionary = []

    #     for dict_word in current_dictionary:
    #         if len(dict_word) != len_word:
    #             continue
    #         if re.match(clean_word, dict_word):
    #             new_dictionary.append(dict_word)

    #     self.current_dictionary = new_dictionary

    #     if len(new_dictionary) == 0:
    #         # If no word matches in the training dictionary, default back to ordering of full dictionary
    #         sorted_letter_count = self.full_dictionary_common_letter_sorted
    #     else:
    #         # Get the most common letter in the possible words that hasn't been guessed yet
    #         full_dict_string = "".join(new_dictionary)
    #         c = collections.Counter(full_dict_string)
    #         sorted_letter_count = c.most_common()
        
    #     # Prepare the features and labels for training the decision tree
    #     X, y = self.prepare_training_data(sorted_letter_count)

    #     # Train the decision tree
    #     decision_tree = DecisionTreeClassifier()
    #     decision_tree.fit(X, y)

    #     # Predict the most likely letter to guess based on the current word pattern
    #     pattern_features = self.extract_pattern_features(word)
    #     predicted_letter = decision_tree.predict([pattern_features])[0]

    #     return predicted_letter

    # def extract_pattern_features(self, word):
    #     # Helper function to extract features from the word pattern
    #     features = []
    #     for letter in string.ascii_lowercase:
    #         features.append(word.count(letter))
    #     return features

    # def prepare_training_data(self, sorted_letter_count):
    #     # Helper function to prepare the training data for the decision tree
    #     X = []
    #     y = []
    #     for letter, instance_count in sorted_letter_count:
    #         if letter not in self.guessed_letters:
    #             features = self.extract_pattern_features(self.construct_word_pattern(letter))
    #             X.append(features)
    #             y.append(letter)
    #     return np.array(X), np.array(y)

    # def construct_word_pattern(self, letter):
    #     # Helper function to construct the word pattern with the guessed letter
    #     pattern = ""
    #     for word in self.current_dictionary:
    #         if letter in word:
    #             pattern += letter
    #         else:
    #             pattern += "_"
    #     return pattern
    #############################################################################################################################################

#############################################################################################################################################
        
        
        
        ######################################################################### ## function 9 (12%)
        
        # # Clean the word by replacing "_" with "."
        # clean_word = word.replace("_", ".")

        # # Find length of the passed word
        # len_word = len(word)  # Use the length of the original word

        # # Grab the current dictionary of possible words from self object
        # current_dictionary = self.current_dictionary
        # new_dictionary = []

        # # Iterate through all of the words in the old plausible dictionary
        # for dict_word in current_dictionary:
        #     # Continue if the word is not of the appropriate length
        #     if len(dict_word) != len_word:
        #         continue

        #     # If dictionary word is a possible match then add it to the current dictionary
        #     if re.match(clean_word, dict_word):
        #         new_dictionary.append(dict_word)

        # # Overwrite old possible words dictionary with the updated version
        # self.current_dictionary = new_dictionary

        # # Check if there are 2 or fewer missing letters
        # num_missing_letters = self.guessed_letters.count("_")
        # if num_missing_letters <= 2:
        #     best_matching_word = None
        #     best_matching_score = 0

        #     # Find a word in the current dictionary that matches the guessed letters so far
        #     guessed_word = ''.join(self.guessed_letters)
        #     for dict_word in new_dictionary:
        #         matching_score = sum(1 for i in range(len_word) if guessed_word[i] == dict_word[i] or guessed_word[i] == '_')
        #         if matching_score > best_matching_score:
        #             best_matching_word = dict_word
        #             best_matching_score = matching_score

        #     if best_matching_word:
        #         guess_letter = [best_matching_word[i] for i in range(len_word) if guessed_word[i] == '_'][0]
        #         return guess_letter

        # # Count the occurrence of all characters in possible word matches
        # full_dict_string = "".join(new_dictionary)
        # letter_count = collections.Counter(full_dict_string)

        # # Sort the letters based on their occurrence frequency in the possible words
        # sorted_letter_count = letter_count.most_common()

        # # Get the most frequent letter in the remaining possible words
        # guess_letter = '!'
        # for letter, instance_count in sorted_letter_count:
        #     guess_letter = letter
        #     break

        # # If no word matches in the training dictionary, default back to the ordering of the full dictionary
        # if guess_letter == '!':
        #     full_dict_letter_count = collections.Counter("".join(self.full_dictionary))
        #     full_dict_sorted_letter_count = full_dict_letter_count.most_common()
        #     for letter, instance_count in full_dict_sorted_letter_count:
        #         if letter not in self.guessed_letters:
        #             guess_letter = letter
        #             break

        # # Update the guessed letters for positions where the guessed letter occurs
        # for i in range(len(word)):
        #     if word[i] == guess_letter:
        #         self.guessed_letters[i] = guess_letter

        # return guess_letter
        
        
        
        
        
        ######################################################################### ## efficiency 12.1%
        # # Clean the word by replacing "_" with "." and removing spaces     
        # clean_word = word.replace("_", "").replace(" ", "")
        
        # # Find length of the passed word
        # len_word = len(clean_word)
        
        # # Grab the current dictionary of possible words from self object
        # current_dictionary = self.current_dictionary
        # new_dictionary = []
        
        # # Iterate through all of the words in the old plausible dictionary
        # for dict_word in current_dictionary:
        #     # Continue if the word is not of the appropriate length
        #     if len(dict_word) != len_word:
        #         continue
            
        #     # If dictionary word is a possible match then add it to the current dictionary
        #     if re.match(clean_word, dict_word):
        #         new_dictionary.append(dict_word)
        
        # # Overwrite old possible words dictionary with updated version
        # self.current_dictionary = new_dictionary
        
        # # Check if there are 2 or fewer missing letters
        # num_missing_letters = self.guessed_letters.count("_")
        # if num_missing_letters <= 2:
        #     best_matching_word = None
        #     best_matching_score = 0

        #     # Find a word in the current dictionary that matches the guessed letters so far
        #     guessed_word = ''.join(self.guessed_letters)
        #     for dict_word in new_dictionary:
        #         matching_score = sum(1 for i in range(len_word) if guessed_word[i] == dict_word[i] or guessed_word[i] == '_')
        #         if matching_score > best_matching_score:
        #             best_matching_word = dict_word
        #             best_matching_score = matching_score

        #     if best_matching_word:
        #         guess_letter = [best_matching_word[i] for i in range(len_word) if guessed_word[i] == '_'][0]
        #         return guess_letter
        
        # # Count occurrence of all characters in possible word matches
        # full_dict_string = "".join(new_dictionary)
        # letter_count = collections.Counter(full_dict_string)
        # sorted_letter_count = letter_count.most_common()
        
        # # Get the most frequent letter in the remaining possible words
        # guess_letter = '!'
        # for letter, instance_count in sorted_letter_count:
        #     if letter not in self.guessed_letters:
        #         guess_letter = letter
        #         break
        
        # # If no word matches in the training dictionary, default back to ordering of full dictionary
        # if guess_letter == '!':
        #     sorted_letter_count = self.full_dictionary_common_letter_sorted
        #     for letter, instance_count in sorted_letter_count:
        #         if letter not in self.guessed_letters:
        #             guess_letter = letter
        #             break
        
        # # Update the guessed letters for positions where the guessed letter occurs
        # for i in range(len(word)):
        #     if word[i] == guess_letter:
        #         self.guessed_letters[i] = guess_letter
        
        # return guess_letter
        
        
        
        
        
        ######################################################################### ## efficiency 12.1%
        # # Clean the word by replacing "_" with "." and removing spaces
        # clean_word = word.replace("_", "").replace(" ", "")
        
        # # Find length of the passed word
        # len_word = len(clean_word)
        
        # # Grab the current dictionary of possible words from self object
        # current_dictionary = self.current_dictionary
        # new_dictionary = []
        
        # # Iterate through all of the words in the old plausible dictionary
        # for dict_word in current_dictionary:
        #     # Continue if the word is not of the appropriate length
        #     if len(dict_word) != len_word:
        #         continue
            
        #     # If dictionary word is a possible match then add it to the current dictionary
        #     if re.match(clean_word, dict_word):
        #         new_dictionary.append(dict_word)
        
        # # Overwrite old possible words dictionary with updated version
        # self.current_dictionary = new_dictionary
        
        # # Check if there are 2 or fewer missing letters
        # num_missing_letters = self.guessed_letters.count("_")
        # if num_missing_letters <= 2:
        #     best_matching_word = None
        #     best_matching_score = 0

        #     # Find a word in the current dictionary that matches the guessed letters so far
        #     guessed_word = ''.join(self.guessed_letters)
        #     for dict_word in new_dictionary:
        #         matching_score = sum(1 for i in range(len_word) if guessed_word[i] == dict_word[i] or guessed_word[i] == '_')
        #         if matching_score > best_matching_score:
        #             best_matching_word = dict_word
        #             best_matching_score = matching_score

        #     if best_matching_word:
        #         guess_letter = [best_matching_word[i] for i in range(len_word) if guessed_word[i] == '_'][0]
        #         return guess_letter
        
        # # Count occurrence of all characters in possible word matches
        # full_dict_string = "".join(new_dictionary)
        # letter_count = collections.Counter(full_dict_string)
        # sorted_letter_count = letter_count.most_common()
        
        # # Consider the positions of the guessed letters and prioritize the letters that are likely to appear in those positions
        # possible_letters = set()
        # for i in range(len_word):
        #     if word[i] == '_':
        #         possible_letters.update({dict_word[i] for dict_word in new_dictionary})
        
        # # Find the most frequent letter from possible letters
        # guess_letter = '!'
        # for letter, instance_count in sorted_letter_count:
        #     if letter not in self.guessed_letters and letter in possible_letters:
        #         guess_letter = letter
        #         break
        
        # # If no word matches in the training dictionary, default back to ordering of full dictionary
        # if guess_letter == '!':
        #     sorted_letter_count = self.full_dictionary_common_letter_sorted
        #     for letter, instance_count in sorted_letter_count:
        #         if letter not in self.guessed_letters:
        #             guess_letter = letter
        #             break
        
        # # Update the guessed letters for positions where the guessed letter occurs
        # for i in range(len(word)):
        #     if word[i] == guess_letter:
        #         self.guessed_letters[i] = guess_letter
        
        # return guess_letter
        
        
        
        
        
        
        # ######################################################################### ## efficiency 12.3% in 25 runs ... 12.5 in 50 runs...   in 100 runs
        # Clean the word by replacing "_" with "."
        clean_word = word.replace("_", ".")
        
        # Find length of the passed word
        len_word = len(word)  # Use the length of the original word
        
        # Grab the current dictionary of possible words from self object
        current_dictionary = self.current_dictionary
        new_dictionary = []
        
        # Iterate through all of the words in the old plausible dictionary
        for dict_word in current_dictionary:
            # Continue if the word is not of the appropriate length
            if len(dict_word) != len_word:
                continue
            
            # If dictionary word is a possible match then add it to the current dictionary
            if re.match(clean_word, dict_word):
                new_dictionary.append(dict_word)
        
        # Overwrite old possible words dictionary with the updated version
        self.current_dictionary = new_dictionary
        
        # Initialize guess_letter before the conditional checks
        guess_letter = '!'
        
        # Check if there are 2 or fewer missing letters
        num_missing_letters = self.guessed_letters.count("_")
        if num_missing_letters <= 2:
            # Find a word in the current dictionary that matches the guessed letters so far
            guessed_word = ''.join(self.guessed_letters)
            best_matching_word = None
            best_matching_score = 0

            for dict_word in new_dictionary:
                matching_score = sum(1 for i in range(len_word) if guessed_word[i] == dict_word[i] or guessed_word[i] == '_')
                if matching_score > best_matching_score:
                    best_matching_word = dict_word
                    best_matching_score = matching_score

            if best_matching_word:
                guess_indices = [i for i in range(len_word) if guessed_word[i] == '_']
                guessed_word = list(guessed_word)
                for index in guess_indices:
                    guessed_word[index] = best_matching_word[index]
                guess_word = ''.join(guessed_word)
                
                # Filter the dictionary based on the guessed word pattern
                self.current_dictionary = [dict_word for dict_word in new_dictionary if re.match(guess_word, dict_word)]
                
                # Sort the letters based on their occurrence frequency in the possible words
                full_dict_string = "".join(self.current_dictionary)
                letter_count = collections.Counter(full_dict_string)
                sorted_letter_count = letter_count.most_common()
                
                # Get the most frequent letter in the remaining possible words
                guess_letter = '!'
                for letter, instance_count in sorted_letter_count:
                    if letter not in self.guessed_letters:
                        guess_letter = letter
                        break
                return guess_letter
            
        
        # If guessed letter is not present in the word, remove words containing that letter from the dictionary
        if guess_letter in word:
            self.current_dictionary = [dict_word for dict_word in new_dictionary if guess_letter not in dict_word]
        
        # Count the occurrence of all characters in possible word matches
        full_dict_string = "".join(new_dictionary)
        letter_count = collections.Counter(full_dict_string)
        
        # Sort the letters based on their occurrence frequency in the possible words
        sorted_letter_count = letter_count.most_common()
        
        # Get the most frequent letter in the remaining possible words
        for letter, instance_count in sorted_letter_count:
            if letter not in self.guessed_letters:
                guess_letter = letter
                break
        
        # If no word matches in the training dictionary, default back to the ordering of the full dictionary
        if guess_letter == '!':
            full_dict_letter_count = collections.Counter("".join(self.full_dictionary))
            full_dict_sorted_letter_count = full_dict_letter_count.most_common()
            for letter, instance_count in full_dict_sorted_letter_count:
                if letter not in self.guessed_letters:
                    guess_letter = letter
                    break
        
        # Update the guessed letters for positions where the guessed letter occurs
        for i in range(len(word)):
            if word[i] == guess_letter:
                self.guessed_letters[i] = guess_letter
        
        return guess_letter
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        ######################################################################### ## efficiency 12.4% in 25 runs....  12.3% in 50 runs...  12.4% in 100 runs
        # # Clean the word by replacing "_" with "."
        # clean_word = word.replace("_", ".")
        
        # # Find length of the passed word
        # len_word = len(word)  # Use the length of the original word
        
        # # Grab the current dictionary of possible words from self object
        # current_dictionary = self.current_dictionary
        # new_dictionary = []
        
        # # Iterate through all of the words in the old plausible dictionary
        # for dict_word in current_dictionary:
        #     # Continue if the word is not of the appropriate length
        #     if len(dict_word) != len_word:
        #         continue
            
        #     # If dictionary word is a possible match then add it to the current dictionary
        #     if re.match(clean_word, dict_word):
        #         new_dictionary.append(dict_word)
        
        # # Overwrite old possible words dictionary with the updated version
        # self.current_dictionary = new_dictionary
        
        # # Initialize guess_letter before the conditional checks
        # guess_letter = '!'
        
        # # Check if there are 2 or fewer missing letters
        # num_missing_letters = self.guessed_letters.count("_")
        # if num_missing_letters <= 2:
        #     # Find a word in the current dictionary that matches the guessed letters so far
        #     guessed_word = ''.join(self.guessed_letters)
        #     best_matching_word = None
        #     best_matching_score = 0

        #     for dict_word in new_dictionary:
        #         matching_score = sum(1 for i in range(len_word) if guessed_word[i] == dict_word[i] or guessed_word[i] == '_')
        #         if matching_score > best_matching_score:
        #             best_matching_word = dict_word
        #             best_matching_score = matching_score

        #     if best_matching_word:
        #         guess_indices = [i for i in range(len_word) if guessed_word[i] == '_']
        #         guessed_word = list(guessed_word)
        #         for index in guess_indices:
        #             guessed_word[index] = best_matching_word[index]
        #         guess_word = ''.join(guessed_word)
                
        #         # Advanced pattern matching
        #         possible_patterns = set()
        #         for i in range(len_word):
        #             if guess_word[i] != '_':
        #                 possible_patterns.add(guess_word[:i] + '.' + guess_word[i+1:])
        #         possible_words = [word for word in new_dictionary if any(re.match(pattern, word) for pattern in possible_patterns)]
                
        #         # Filter the dictionary based on advanced pattern matching
        #         self.current_dictionary = possible_words
                
        #         # Sort the letters based on their occurrence frequency in the possible words
        #         full_dict_string = "".join(self.current_dictionary)
        #         letter_count = collections.Counter(full_dict_string)
        #         sorted_letter_count = letter_count.most_common()
                
        #         # Get the most frequent letter in the remaining possible words
        #         guess_letter = '!'
        #         for letter, instance_count in sorted_letter_count:
        #             if letter not in self.guessed_letters:
        #                 guess_letter = letter
        #                 break
        #         return guess_letter
                
        
        # # If guessed letter is not present in the word, remove words containing that letter from the dictionary
        # if guess_letter in word:
        #     self.current_dictionary = [dict_word for dict_word in new_dictionary if guess_letter not in dict_word]
        
        # # Count the occurrence of all characters in possible word matches
        # full_dict_string = "".join(new_dictionary)
        # letter_count = collections.Counter(full_dict_string)
        
        # # Sort the letters based on their occurrence frequency in the possible words
        # sorted_letter_count = letter_count.most_common()
        
        # # Get the most frequent letter in the remaining possible words
        # for letter, instance_count in sorted_letter_count:
        #     if letter not in self.guessed_letters:
        #         guess_letter = letter
        #         break
        
        # # If no word matches in the training dictionary, default back to the ordering of the full dictionary
        # if guess_letter == '!':
        #     full_dict_letter_count = collections.Counter("".join(self.full_dictionary))
        #     full_dict_sorted_letter_count = full_dict_letter_count.most_common()
        #     for letter, instance_count in full_dict_sorted_letter_count:
        #         if letter not in self.guessed_letters:
        #             guess_letter = letter
        #             break
        
        # # Update the guessed letters for positions where the guessed letter occurs
        # for i in range(len(word)):
        #     if word[i] == guess_letter:
        #         self.guessed_letters[i] = guess_letter
        
        # return guess_letter







        
        
        
        
        
        
        
        







        
        
        
        


    
        #############################################################################################################################################
        
        



        # # clean the word so that we strip away the space characters
        # # replace "_" with "." as "." indicates any character in regular expressions
        # clean_word = word[::2].replace("_",".")
        
        # # find length of passed word
        # len_word = len(clean_word)
        
        # # grab current dictionary of possible words from self object, initialize new possible words dictionary to empty
        # current_dictionary = self.current_dictionary
        # new_dictionary = []
        
        # # iterate through all of the words in the old plausible dictionary
        # for dict_word in current_dictionary:
        #     # continue if the word is not of the appropriate length
        #     if len(dict_word) != len_word:
        #         continue
                
        #     # if dictionary word is a possible match then add it to the current dictionary
        #     if re.match(clean_word,dict_word):
        #         new_dictionary.append(dict_word)
        
        # # overwrite old possible words dictionary with updated version
        # self.current_dictionary = new_dictionary
        
        
        # # count occurrence of all characters in possible word matches
        # full_dict_string = "".join(new_dictionary)
        
        # c = collections.Counter(full_dict_string)
        # sorted_letter_count = c.most_common()                   
        
        # guess_letter = '!'
        
        # # return most frequently occurring letter in all possible words that hasn't been guessed yet
        # for letter,instance_count in sorted_letter_count:
        #     if letter not in self.guessed_letters:
        #         guess_letter = letter
        #         break
            
        # # if no word matches in training dictionary, default back to ordering of full dictionary
        # if guess_letter == '!':
        #     sorted_letter_count = self.full_dictionary_common_letter_sorted
        #     for letter,instance_count in sorted_letter_count:
        #         if letter not in self.guessed_letters:
        #             guess_letter = letter
        #             break            
        
        # return guess_letter

        

    ##########################################################
    # You'll likely not need to modify any of the code below #
    ##########################################################
    
    def build_dictionary(self, dictionary_file_location):
        text_file = open(dictionary_file_location,"r")
        full_dictionary = text_file.read().splitlines()
        text_file.close()
        return full_dictionary
                
    def start_game(self, practice=True, verbose=True):
        # reset guessed letters to empty set and current plausible dictionary to the full dictionary
        self.guessed_letters = []
        self.current_dictionary = self.full_dictionary
                         
        response = self.request("/new_game", {"practice":practice})
        if response.get('status')=="approved":
            game_id = response.get('game_id')
            word = response.get('word')
            tries_remains = response.get('tries_remains')
            if verbose:
                print("Successfully start a new game! Game ID: {0}. # of tries remaining: {1}. Word: {2}.".format(game_id, tries_remains, word))
            while tries_remains>0:
                # get guessed letter from user code
                guess_letter = self.guess(word)
                    
                # append guessed letter to guessed letters field in hangman object
                self.guessed_letters.append(guess_letter)
                if verbose:
                    print("Guessing letter: {0}".format(guess_letter))
                    
                try:    
                    res = self.request("/guess_letter", {"request":"guess_letter", "game_id":game_id, "letter":guess_letter})
                except HangmanAPIError:
                    print('HangmanAPIError exception caught on request.')
                    continue
                except Exception as e:
                    print('Other exception caught on request.')
                    raise e
               
                if verbose:
                    print("Sever response: {0}".format(res))
                status = res.get('status')
                tries_remains = res.get('tries_remains')
                if status=="success":
                    if verbose:
                        print("Successfully finished game: {0}".format(game_id))
                    return True
                elif status=="failed":
                    reason = res.get('reason', '# of tries exceeded!')
                    if verbose:
                        print("Failed game: {0}. Because of: {1}".format(game_id, reason))
                    return False
                elif status=="ongoing":
                    word = res.get('word')
        else:
            if verbose:
                print("Failed to start a new game")
        return status=="success"
        
    def my_status(self):
        return self.request("/my_status", {})
    
    def request(
            self, path, args=None, post_args=None, method=None):
        if args is None:
            args = dict()
        if post_args is not None:
            method = "POST"

        # Add `access_token` to post_args or args if it has not already been
        # included.
        if self.access_token:
            # If post_args exists, we assume that args either does not exists
            # or it does not need `access_token`.
            if post_args and "access_token" not in post_args:
                post_args["access_token"] = self.access_token
            elif "access_token" not in args:
                args["access_token"] = self.access_token

        time.sleep(0.2)

        num_retry, time_sleep = 50, 2
        for it in range(num_retry):
            try:
                response = self.session.request(
                    method or "GET",
                    self.hangman_url + path,
                    timeout=self.timeout,
                    params=args,
                    data=post_args,
                    verify=False
                )
                break
            except requests.HTTPError as e:
                response = json.loads(e.read())
                raise HangmanAPIError(response)
            except requests.exceptions.SSLError as e:
                if it + 1 == num_retry:
                    raise
                time.sleep(time_sleep)

        headers = response.headers
        if 'json' in headers['content-type']:
            result = response.json()
        elif "access_token" in parse_qs(response.text):
            query_str = parse_qs(response.text)
            if "access_token" in query_str:
                result = {"access_token": query_str["access_token"][0]}
                if "expires" in query_str:
                    result["expires"] = query_str["expires"][0]
            else:
                raise HangmanAPIError(response.json())
        else:
            raise HangmanAPIError('Maintype was not text, or querystring')

        if result and isinstance(result, dict) and result.get("error"):
            raise HangmanAPIError(result)
        return result
    
class HangmanAPIError(Exception):
    def __init__(self, result):
        self.result = result
        self.code = None
        try:
            self.type = result["error_code"]
        except (KeyError, TypeError):
            self.type = ""

        try:
            self.message = result["error_description"]
        except (KeyError, TypeError):
            try:
                self.message = result["error"]["message"]
                self.code = result["error"].get("code")
                if not self.type:
                    self.type = result["error"].get("type", "")
            except (KeyError, TypeError):
                try:
                    self.message = result["error_msg"]
                except (KeyError, TypeError):
                    self.message = result

        Exception.__init__(self, self.message)
```
